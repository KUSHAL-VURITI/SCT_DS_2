```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Titanic_data.csv')

data['Age'] = data['Age'].fillna(data['Age'].median())
data['Fare'] = data['Fare'].fillna(data['Fare'].median())
data = data.drop(columns=['Cabin'])

fig, axes = plt.subplots(2, 2, figsize=(20, 10))

sns.countplot(data=data, x='Survived', hue='Survived', palette='pastel', ax=axes[0, 0])
axes[0, 0].set_title('Survival Distribution (0=Not Survived, 1=Survived)')

sns.barplot(data=data, x='Sex', y='Survived', hue='Sex', palette='coolwarm', ax=axes[0, 1])
axes[0, 1].set_title('Survival rate by Gender')

sns.barplot(data=data, x='Pclass', y='Survived', hue='Pclass', palette='muted', ax=axes[1, 0])
axes[1, 0].set_title('Survival rate by Passenger class')

sns.histplot(data=data, x='Age', hue='Survived', kde=True, palette='Set1', bins=30, ax=axes[1, 1])
axes[1, 1].set_title('Age Distribution by survival')

plt.tight_layout()
plt.show()
```
